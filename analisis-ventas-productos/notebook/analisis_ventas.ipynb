```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/ventas.csv")
df.head()

df["ingreso"] = df["precio"] * df["cantidad"]
df["costo_total"] = df["costo_unitario"] * df["cantidad"]
df["utilidad"] = df["ingreso"] - df["costo_total"]

resumen = df.groupby("producto")[["ingreso","utilidad"]].sum()
print(resumen)

resumen["margen"] = resumen["utilidad"] / resumen["ingreso"]
print(resumen)

plt.bar(resumen.index, resumen["ingreso"])
plt.title("Ingresos por producto")
plt.xlabel("Producto")
plt.ylabel("Ingresos")
plt.show()

plt.bar(resumen.index, resumen["utilidad"])
plt.title("Utilidad por producto")
plt.xlabel("Producto")
plt.ylabel("Utilidad")
plt.show()

plt.bar(resumen.index, resumen["margen"])
plt.title("Margen por producto")
plt.xlabel("Producto")
plt.ylabel("Margen")
plt.show()
```
